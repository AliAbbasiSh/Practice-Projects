```python
import pandas_datareader as pdd
import pandas as pd
import numpy as np
import seaborn as sns
import datetime
%matplotlib inline
```

```python
start = datetime.datetime(2006,1,1)
end = datetime.datetime(2016,1,1)
```

```python
BAC = pdd.DataReader("BAC", 'yahoo', start, end)
C = pdd.DataReader("C", 'yahoo', start, end)
GS = pdd.DataReader("GS", 'yahoo', start, end)
JPM = pdd.DataReader("JPM", 'yahoo', start, end)
MS = pdd.DataReader("MS", 'yahoo', start, end)
WFC = pdd.DataReader("WFC", 'yahoo', start, end)
```

```python
MS
```

```python
tickers=['BAC','C','GS','JPM','MS','WFC']
```

```python
bank_stocks=pd.concat([BAC,C,GS,JPM,MS,WFC],axis=1,keys=tickers)
bank_stocks
```

```python
bank_stocks.columns.names = ['Bank Ticker','Stock Info']
```

```python
bank_stocks
```

```python
bank_stocks.xs(key="Close",axis=1,level="Stock Info").max()
```

```python
for tick in tickers:
    print(tick,bank_stocks[tick]['Close'].max())
```

```python
returns = pd.DataFrame()
```

```python
for tick in tickers:
    returns[tick + "Return"]=bank_stocks[tick]["Close"].pct_change()
returns
```
